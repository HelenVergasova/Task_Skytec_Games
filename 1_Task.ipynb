#1  Успешность прототипа


Известно что компания конкурент выпустила 1000 прототипов игр, из которых было 5 успешных. Наша компания выпустила 200 прототипов из которых ни один не был успешен.
Какова вероятность что следующий (201й) наш прототип будет успешен? Решите задачу классическим и баесовским подходами.

Классический подход:

Для нашей компании:

Вероятность успеха=успешные прототипы/общее количество прототипов=0/200=0

Для компании конкурента:

Вероятность успеха=успешные прототипы/общее количество прототипов=5/1000=0.005

Итак, если принять, что вероятность успешности прототипа не зависит от предыдущих результатов, то вероятность успешности следующего прототипа для нашей компании также будет 0.005, так как мы не обладаем информацией о том, почему, предыдущиее прототипы былии неуспешными.

Баесовский подход:

В баесовском подходе мы учитываем априорную информацию о вероятности успешности и обновляем ее с учетом новых данных.

Пусть:
P(A) - априорная вероятность успешности нашего прототипа.
P(B|A)-вероятность того, чтоо новый прототип будет успешным, при условии, что априорная вероятность успешности А.
P(B) - общая вероятность успешности нового прототипа.

Известно:
P(A) = 0, так как не имеем информации о предыдущих прототипах нашей компании.
P(B|A) = 0.005, так как это вероятность успешности конкурентной компании.

Используем формулу Байеса:
P(A|B)=(P(B|A)*P(A))/(P(B))

P(B) = P(B|A)*P(A)+P(B|-A)*P(-A)
Здесь: P(-A)=1-P(A)

Теперь вычислим P(B) и P(A|B):
P(B)=0.005*0+P(B|-A)*(1-0)

Так как мы не имеем информации о вероятности успешности нашей компании, мы нне можем вычислить P(B|-A).
Таким образом, мы не можем решить задачу баесовским подхоодом без дополнительной информации о нашей компании.
В общих подходах мы приходим к выводу, что без дополнительной иннформации о предыдущих прототипах нашей компании невозмможно точно предскказать вероятноость успешности следующего прототипа.
